```python
# divide the dataset into independent and dependent
X=df.iloc[:,:-1]
Y=df.iloc[:,-1]
```

```python
X.shape, Y.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.33, random_state=36)
```

```python
X_train.shape, X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
# scaler.fit(X_train)
# scaler.transform(X_train)
X_train=scaler.fit_transform(X_train)
```

```python
X_train
```

```python
X_test=scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression.fit(X_train,y_train)
```

```python
regression.coef_
```

```python
regression.intercept_
```

```python
## Predicton
y_pred=regression.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
import numpy as np
mse=mean_squared_error(y_test,y_pred)
print(mse)
mae=mean_absolute_error(y_test,y_pred)
print(mae)
print(np.sqrt(mse))
```

```python
# Accuracy r2 and adjusted r square
from sklearn.metrics import r2_score
```

```python
score=r2_score(y_test,y_pred)
score
```

```python
#display adjusted R-squared
1 - (1-score)*(len(Y)-1)/(len(Y)-X.shape[1]-1)
```

```python
from sklearn.linear_model import Ridge
ridge=Ridge()#alpha=20
ridge.fit(X_train,y_train)
```

```python
y_pred=ridge.predict(X_test)
```

```python
mse=mean_squared_error(y_test,y_pred)
print(mse)
mae=mean_absolute_error(y_test,y_pred)
print(mae)
print(np.sqrt(mse))
```

```python
from sklearn.linear_model import Lasso
lasso=Lasso()#alpha=20
lasso.fit(X_train,y_train)
```

```python
y_pred=lasso.predict(X_test)
mse=mean_squared_error(y_test,y_pred)
print(mse)
mae=mean_absolute_error(y_test,y_pred)
print(mae)
print(np.sqrt(mse))
```

```python
from sklearn.linear_model import ElasticNet
elasticnet=ElasticNet()#alpha=20
elasticnet.fit(X_train,y_train)
```

```python
y_pred=elasticnet.predict(X_test)
mse=mean_squared_error(y_test,y_pred)
print(mse)
mae=mean_absolute_error(y_test,y_pred)
print(mae)
print(np.sqrt(mse))
```
